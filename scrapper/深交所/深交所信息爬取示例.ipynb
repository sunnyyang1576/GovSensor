```python
HEADER = {'User-Agent':"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36"}
```

```python
def query_URL_construction(catelogid,tabkey,pageno,random=0.15390417443881077):
    
    base="http://www.szse.cn/api/report/ShowReport/data?"
    
    URL = base+"SHOWTYPE=JSON"+"&CATALOGID="+str(catelogid)+"&TABKEY=tab"+str(tabkey)+"&PAGENO="+str(pageno)+"&random="+str(random)
    
    return URL
    
    
    
    
```
